```python
def read_tskv(filepath: str) -> Optional[pd.DataFrame]:
    """
    Функция для чтения датасета по указанному пути

    args:
    filepath (str) - путь к файлу .pqt / .parquet

    return:
    датасет в формате pd.DataFrame, если найден файл по указанному пути
    """
    if os.path.exists(filepath):
        try:
            list_file = []
            with open('geo-reviews-dataset-2023.tskv', 'r', encoding='utf-8') as file:
                lines = file.readlines()
                for line in lines:
                    line = line.split('\t')
                    dict_row = {}
                    for l in line:
                        l = l.split('=')
                        dict_row[l[0]] = l[1]
                    list_file.append(dict_row)
            return pd.DataFrame(list_file)
        except error as error:
            print(error)
    print(f"Файл не найден по пути {filepath}")
    return
```

### Лемматизации

```python
morph = pymorphy2.MorphAnalyzer()


def lemmatize_text(text):
    words = text.split()
    lemmatized_words = [morph.parse(word)[0].normal_form for word in words]
    lemmatized_text = " ".join(lemmatized_words)
    return lemmatized_text
```

```python
tqdm.pandas()

df['text'] = df['text'].progress_apply(lemmatize_text)
```

### Баланс классов

У нас большой датасет и минорный класс имеет `2` имеет 12071 отзыв, так что я возьму у каждого класса по 10000 отызвов.

```python
def make_autopct(values):
    def my_autopct(pct):
        total = sum(values)
        val = int(round(pct*total/100.0))
        return '{p:.2f}%  ({v:d})'.format(p=pct,v=val)
    return my_autopct
```

```python
rating_counts = df['rating'].value_counts()


plt.figure(figsize=(4, 4))
plt.pie(rating_counts, labels=rating_counts.index, autopct= make_autopct(rating_counts) , startangle=30)
plt.axis('equal')
plt.title('Распределение отзывов по оценке')

plt.show()
```

```python
class_1 = df[df['rating'] == 1]
class_2 = df[df['rating'] == 2]
class_3 = df[df['rating'] == 3]
class_4 = df[df['rating'] == 4]
class_5 = df[df['rating'] == 5]


size = 50000



class_1_undersampled = resample(class_1, replace=True, n_samples=size)
class_2_undersampled = resample(class_2, replace=True, n_samples=size)
class_3_undersampled = resample(class_3, replace=True, n_samples=size)
class_4_undersampled = resample(class_4, replace=True, n_samples=size)
class_5_undersampled = resample(class_5, replace=False, n_samples=size)


df = pd.concat([class_1_undersampled, class_2_undersampled, class_3_undersampled, class_4_undersampled, class_5_undersampled])


rating_counts = df['rating'].value_counts()


plt.figure(figsize=(4, 4))
plt.pie(rating_counts, labels=rating_counts.index, autopct= make_autopct(rating_counts) , startangle=30)
plt.axis('equal')
plt.title('Распределение отзывов по оценке')

plt.show()
```

Я пробывал два способа векторизации. tfidf - обучался быстрее и показывал лучше точность. 

```python
# cv=CountVectorizer(min_df=0,max_df=0.95 ,binary=False,ngram_range=(1,3))

# X_train_bow =cv.fit_transform(X_train)

# X_test_bow=cv.transform(X_test)

# print('BOW_cv_train:',X_train_bow.shape)
# print('BOW_cv_test:',X_test_bow.shape)
```

Здесь я использую параметры:

```python
tfidf_vectorizer = TfidfVectorizer(
    min_df=1, binary=False, ngram_range=(1, 3))

X_train.fillna('', inplace=True)
X_test.fillna('', inplace=True)


X_train_tfidf = tfidf_vectorizer.fit_transform(X_train)

X_test_tfidf = tfidf_vectorizer.transform(X_test)


print('Tfidf_train:', X_train_tfidf.shape)
print('Tfidf_test:', X_test_tfidf.shape)
```
